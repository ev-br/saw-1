```python

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)

counts = np.loadtxt("Homopolymer_simulations/Additional_1/counts_isaw_moresteps_0.000000_higher_hpc.txt", skiprows=6, dtype=int) 
sums = np.zeros(np.max(counts[:,0])+1)
for i in counts:
    sums[i[0]] = sums[i[0]] + i[2]
plt.plot(np.arange(np.max(counts[:,0])+1),  sums/np.sum(counts[:,2]), "s", c = "pink", lw=9, label = r"$\beta=0.370093$ ($\langle N \rangle \approx 55.2262$)")


counts = np.loadtxt("Homopolymer_simulations/Main_simulations/counts_isaw_moresteps_0.000000_higher_hpc.txt", skiprows=6, dtype=int) 
sums = np.zeros(np.max(counts[:,0])+1)
for i in counts:
    sums[i[0]] = sums[i[0]] + i[2]
plt.plot(np.arange(np.max(counts[:,0])+1),  sums/np.sum(counts[:,2]), "o", c = "purple", lw=9, label = r"$\beta=0.376438$ ($\langle N \rangle \approx 193.019$)")

counts = np.loadtxt("Homopolymer_simulations/Additional_2/counts_isaw_moresteps_0.000000_higher_hpc.txt", skiprows=6, dtype=int) 
sums = np.zeros(np.max(counts[:,0])+1)
for i in counts:
    sums[i[0]] = sums[i[0]] + i[2]
plt.plot(np.arange(np.max(counts[:,0])+1),  sums/np.sum(counts[:,2]), "D", c = "tomato", lw=9, label = r"$\beta=0.377602$ ($\langle N \rangle \approx 347.978$)")

plt.xlim(0, 1000)

plt.tick_params(axis='x', labelsize=20) 
plt.tick_params(axis='y', labelsize=20) 

plt.title("Density of length", fontsize=20)
#ax.set_yscale('log')
#ax.set_xscale('log')

plt.legend(loc="best", fontsize=20)

plt.savefig("/home/kamilla/SAWs/Bachelor thesis/img/densities_saw_1.png")
```
