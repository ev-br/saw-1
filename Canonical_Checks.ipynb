```python
import matplotlib.pyplot as plt
import random
import numpy as np

import pandas as pd
from matplotlib.ticker import LogFormatter 
from scipy.optimize import curve_fit
from pathlib import Path
import matplotlib.colors
from matplotlib import ticker, cm
import pickle
```

```python

with open('Ising/radiues10.5.pickle', 'rb') as f:
    x = pickle.load(f)

fig = plt.figure(figsize=(8, 7))
ax = fig.add_subplot(111)

results = np.loadtxt("Ising/Calcs_1/Radius_0.500000_0_higher_hpc.txt", skiprows=3)
n_saw = results[:,0]
mean_r = results[:,1]

plt.title(r"End-to-end distance $R_e^2$", fontsize=24)
plt.errorbar( n_saw[:13], mean_r[:13] , yerr=results[:13,2], fmt="o--", alpha=0.5,ms=10, c="green",label =  "BC data")

plt.plot(np.arange(4, 10), x, "s--",c = "b", ms=14, alpha=0.4 , label = "Exact enumeration")




Canonical_data = np.loadtxt("Ising/Test_for_canonicalModel/All_0.5.txt")
plt.errorbar(Canonical_data[:,1], Canonical_data[:,4], yerr=Canonical_data[:,5], fmt="s", c = "k", ms=10, alpha=0.95, label = "Canonical")



ax.set_xticks(np.arange(2, 15, 2))

plt.xlabel("N", fontsize = 24)
plt.ylabel( r"$R_e^2$" , fontsize=20)
plt.tick_params(axis='x', labelsize=20) 
plt.tick_params(axis='y', labelsize=20) 


plt.legend(loc="best", fontsize=20)

plt.savefig("/home/kamilla/Lattice_models/img/EeBcCanonical05.png")
```
