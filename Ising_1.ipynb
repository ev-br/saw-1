### Heat capacity

```python
nmin = 200 
nmax = 400

e = 10

data = np.zeros(( nmax-nmin+1, e ))
 
for i in range(e):
    filename = "Ising/Calcs_1/M_counts_and_contacts_"+str(i/10)+"00000_0_higher_hpc.txt"
    filedata = np.loadtxt(filename, skiprows=2) 
    for N in range(nmin, nmax+1):
        currentdata=filedata[filedata[:,0]==N]     
        data[N-nmin, i] = currentdata[:,3] - currentdata[:,1]*currentdata[:,1]
```

```python
x = np.arange(nmin, nmax+1)
y = np.arange(0, e)/10.0
X, Y = np.meshgrid(x, y)


plt.figure(figsize=(10,5.5))
plt.contourf(y, x,  data[:,:] ,extend='both', cmap = "plasma")
#plt.xlabel(r"$\epsilon_b$", fontsize=20)
plt.xlabel(r"J", fontsize=20)
plt.ylabel("N", fontsize=20)
plt.tick_params(axis='x', labelsize=20) 
plt.tick_params(axis='y', labelsize=20) 
plt.colorbar()
plt.title('Heat capacity', fontsize=20)


plt.savefig("/home/kamilla/Lattice_models/img/heatcapacityBS.png")
```

$\nu$

```python
def r_on_n(n, nu, b):
    
    
    return 2*nu*n+b
```

```python
    
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)

 

v = np.zeros(10)
sigmas = np.zeros(10)
ints = np.arange(10)

k1=1
k2=1

nmin = 70
nmax_1 = 401
for i in range(10):
    filename ="Ising/Calcs_1/Radius_"+str(i/10)+"00000_0_higher_hpc.txt"
    results = np.loadtxt(filename, skiprows=2) 
    results=results[nmin:nmax_1,:]
    n_saw = results[:,0]
    mean_r = results[:,1]
    err_r = results[:,2]    
    popt, pcov = curve_fit(r_on_n, np.log(n_saw+k2), np.log(mean_r+k1), sigma=err_r )  
    v[i] = popt[0]
    perr = np.sqrt(np.diag(pcov))

    sigmas[i] =perr[0]

#ax2.plot(ints/10, v , ".", c="k") 
plt.errorbar( ints/10, v , yerr=sigmas, fmt="o", ms = 12,c="k", label = r"Estimated $\nu$")
nu_homopolymer = v
#plt.xlabel(r'$\epsilon_b$', fontsize = 22)
plt.xlabel(r'J', fontsize = 22)
plt.ylabel(r"$\nu$ ", fontsize=22)
plt.tick_params(axis='x', labelsize=22) 
plt.tick_params(axis='y', labelsize=22) 
ax.set_xticks(np.arange(0, 1, 0.1))
 
 
theta = 1.0/1.18

plt.plot([theta], [4.0/7.0], "*", c= "purple", ms=15)

plt.legend(loc="best", fontsize=20)

#plt.grid("True")

ax2 = ax.twiny() 
#ax2.set_xlim(0,1)
ax2.set_xlabel(r'$x$', fontsize=22)
ax2.set_xticks(np.arange(0,10,2)/10)
ax2.tick_params(axis='x', labelsize=22)
ax2.set_xbound(ax.get_xbound())
ax2.set_xticklabels( np.round( np.exp(np.arange(0, 10, 2)/10), 2) )
    
    
fig.tight_layout()


plt.savefig("/home/kamilla/Lattice_models/img/nuvsJBS.png")
```

```python
nplot = 10 
data = np.zeros(( nmax-nmin+1, nplot , 11 ))

#lengths = [100, 150, 200, 250, 300, 350, 400]

#lengths = [100, 150, 200, 250, 300 ]

lengths = [50, 100, 150, 200, 250 ]

datas = np.zeros((len(lengths), nplot ))
errors = np.zeros((len(lengths), nplot ))
for i in range(nplot ):
    filename = "Ising/Calcs_1/M_counts_and_contacts_"+str(i/10)+"00000_0_higher_hpc.txt"
    filedata = np.loadtxt(filename, skiprows=2) 
    
    
    for l in range(len(lengths)):
        current = filedata[lengths[l]] #filedata[filedata[:,0]==lengths[l]]
        datas[l, i] =  current[5] 
        errors[l,i] =  current[6] 
        #datas[l, i] =  current[3] - current[1]*current[1]
        #errors[l,i] = current[4]*current[4] + 2*current[2]*current[2]*current[1]
        errors[l,i] = np.sqrt(errors[l,i])
        
  

    
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)

x = np.arange(nplot)/10
for l in range(len(lengths)):
    plt.errorbar(x, datas[l, :], yerr=errors[l,:], fmt="s", ms = 10, label = "N = " + str(lengths[l]))
    

 
plt.xlabel(r'$\epsilon_b$', fontsize = 22)
plt.ylabel(r"$ \langle M \rangle $ ", fontsize=22)
plt.tick_params(axis='x', labelsize=22) 
plt.tick_params(axis='y', labelsize=22) 
#plt.ylim(-1)
  

plt.legend(loc="best", fontsize=20)
 
ax2 = ax.twiny() 
ax2.set_xlabel(r'$x$', fontsize=22)
ax2.set_xticks(np.arange(0,10,2)/10)
ax2.tick_params(axis='x', labelsize=22)
ax2.set_xbound(ax.get_xbound())
ax2.set_xticklabels( np.round( np.exp(np.arange(0, nplot, 2)/10), 2) )
```
