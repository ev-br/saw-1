```python
N = 150 

filename = "/home/kamilla/SAWs/HP_attempt1/for/counts_mags_hp_moresteps_0."+str(0)+"00000_higher_hpc.txt"
results = np.loadtxt(filename, skiprows=5) 
results=results[results[:,0]==N]
meanm = np.sum(results[:,1]*results[:,2]/N)/np.sum(results[:,2])
meansq = np.sum(np.square(results[:,1])*results[:,2])/np.sum(results[:,2])

print(r"Mean M = ", meanm)

filename = "/home/kamilla/SAWs/HP_attempt1/for/counts_hp_moresteps_0."+str(0)+"00000_higher_hpc.txt"
results_m = np.loadtxt(filename, skiprows=9) 
results_m=results_m[results_m[:,0]==N]
meanm_m = np.sum(results_m[:,1]*results_m[:,2])/np.sum(results_m[:,2])
meansq_m = np.sum(np.square(results_m[:,1])*results_m[:,2])/np.sum(results_m[:,2])

 
```

Так как веса любых последовательностей одинаковы, то и средняя плотность Н должна быть близка к 1/2

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)

plt.plot(results[:,1],  results[:,2]/np.sum(results[:,2]), "D", c = "tomato", lw=9, label = r"$ \epsilon_b = 0.0 $)")

plt.xlabel(r"Число Н в последовательности", fontsize = 20)
plt.ylabel(r"  ", fontsize=20)
plt.tick_params(axis='x', labelsize=20) 
plt.tick_params(axis='y', labelsize=20) 
plt.title('Распределение Н ', fontsize=20)

plt.subplot(1,2,2)
plt.plot(results_m[:,1],  results_m[:,2]/np.sum(results_m[:,2]), "D", c = "tomato", lw=9, label = r"$ \epsilon_b = 0.0 $)")


plt.xlabel(r"Число топологических контактов НН", fontsize = 20)
plt.ylabel(r" ", fontsize=20)
plt.tick_params(axis='x', labelsize=20) 
plt.tick_params(axis='y', labelsize=20) 
plt.title('Распределение контактов НН ', fontsize=20)
```

Здесь все нормально.
