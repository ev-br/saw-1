```python
from math import sqrt 
from math import floor
from math import ceil
from math import exp 


def count_saws_contact(n):
    """ n - число мономеров.
    Максимальное число топологических соседей для длины 1"""
     
    m_1 = ceil(sqrt(n))
    m = floor(sqrt(n))
    if( n>m*m and m*m_1>=n ):
        t_max =n+1 - (4*m+2)/2
    elif (n>m*m_1 and n<=m_1*m_1):
        t_max = n+1 - (4*m+4)/2
    else:
        t_max =  n+1 - (4*m)/2
    
    t_max=int(t_max)
    
    return t_max
```
